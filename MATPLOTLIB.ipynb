***You can also use the shortcut string notation parameter to specify the marker***

***This parameter is also called fmt, and is written with this syntax:***
***marker|line|color***

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:m',ms=20)  #marker size is ms=size 
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:m',ms=20,mec="r")  #marker size is ms=size
#mec for marker edge color
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:m',ms=20,mfc="#DAA520",mec="#DAA520")  #marker size is ms=size
#mfc for marker face color
plt.show()
```

```python
plt.plot(ypoints, "*-",ms=4,mec="r",linewidth=10)
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1,c="r") #for chosing color syntax c="color"
plt.plot(y2)

plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1,c="r")
plt.plot(x2,y2,"o:m")
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y,'o:m')

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Sports watch")
```

```python
font1 = {'family':'serif','color':'blue','size':30} #font descrption 
plt.plot(x, y,'o:m')

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid(axis = 'y')
plt.title("Sports watch",fontdict=font1,loc="right") #right and font descrption 
```
